```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
def convertBGR2RGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
def convertBGR2GRAY(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
linkToPhoto = './gradpic1.png'
gradpic = cv2.imread(linkToPhoto)
imageConvertedToRGB = convertBGR2RGB(gradpic)

# plt.imshow(gradpic, cmap='cv2.COLOR_BGR2RGB')
# cv2.COLOR_BGR2RGB
```

```python
grayscaled = convertBGR2GRAY(gradpic)
plt.imshow(grayscaled, cmap='gray')
```
